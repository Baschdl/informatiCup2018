```python
from ipyleaflet import Map, Marker, Circle, CircleMarker, ImageOverlay

m = Map(center=list(berlin), zoom=10)
mark = Marker(location=m.center)
mark.visible
c = Circle(location=m.center)
m.add_layer(c)

cm = CircleMarker(location=m.center, radius=30, weight=2,
                  color='#000', opacity=1.0, fill_opacity=1.0,
                  fill_color='#0F0')
m.add_layer(cm)
io = ImageOverlay(url='http://ipython.org/_static/IPy_header.png', bounds=m.bounds)
m.add_layer(io)
m
```
