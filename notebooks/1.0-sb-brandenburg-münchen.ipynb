```python
def load_gas_station(id):
    gas_station = pd.read_csv(os.path.join(GAS_PRICE_PATH,'{}.csv'.format(id)), names=['Timestamp', 'Price'],  index_col='Timestamp',parse_dates=['Timestamp'],sep=';')
    return gas_station.resample('1T').mean().bfill() 
```

```python
for id in [gas_station_id_brandenburg, gas_station_id_muenchen, gas_station_id_berlin_agip]:
    gas_station_resampled = load_gas_station(id)
    plt.plot(gas_station_resampled)
    plt.gcf().autofmt_xdate()
    plt.show()
```
